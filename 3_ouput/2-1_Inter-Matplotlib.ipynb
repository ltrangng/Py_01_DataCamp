### **Matplotlib**

There are many data visualization packages in Python, but the mother of all is **matplotlib**. You will need the subpackage `pyplot`. By convention it is imported as `plt`.

```python
import matplotlib.pyplot as plt
```

For example, let's try to gain some insights in the evolution of the world population.

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972] # expressed in billions.
```

To plot this data as a line chart, we call `plt.plot()` and use our two lists as arguments. The first argument corresponds to the horizontal axis, the second to the vertical. 

```python
plt.plot(year, pop) 
plt.show()
```

The plot does not pop up right away. You will have to call the `plot.show()` function first. This is because you might want to add some more extra elements to your plot before actually displaying it, such as title and label customizations.

The `plt.plot()` function tells Python what to plot and how to plot it. `plt.show()` actually displays the plot.

There are four data points, and Python draws a line between them.



Another type of plot is the scatter plot. The resulting plot simply plots all the individual data points.

```python
plt.scatter(year, pop)
plt.show()
```

### **Histogram**

Histogram is a type of visualization that is very useful to explore your data. It shows the distribution of the variables.  
Too se how histogram works, imagine we have 12 values between 0 and 6. To build a histogram for these values, divide the number line into equal chunks called *bins*. Next you count how many data points sit inside each bin. Finally, you draw a bar for each bin. The result is a histogram which shows an overview on how the 12 values are distributed: most values are in the middle, and there are more values below 2 than above 4.




`hist()` function in matplotlib is used to build a histogram. 

`plt.hist(x, bins =...)`


Ask Help for more information:

```python
help(plt.hist)
```

There are a lot of arguments you can specify, but the most import are `x`, which is the list of values you want to build data for, and `bins` to tell Python how many bins the data should be divided. Based on this number,  `hist` will automatically find appropriate boundaries for all bins, and calculate how many values are in each one. By default `bins = 10`.

Generate a histogram with the list of 12 values:

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values, bins = 3)
plt.show()
```

Histograms are really useful to give a bigger picture. As an example, the population pyramid is one form of horizontal bin histogram. The ages distribution is shown, both males and females, in EU. The bins are largest for the ages from 40 to 44 in 2010. But the figures will change in 2050 as the baby boomer will get older. Histogram makes it easy to see how demographics changes over the time.



Create a histogram for `life_exp` to see how life expectancy in different countries is distributed:

```python
plt.hist(life_exp)
plt.show()
```

When you do not specify the number of bins, by default, Python sets the number of bins to 10 in that case. The number of bins is pretty important. Too few bins will oversimplify reality and won't show you the details. Too many bins will overcomplicate reality and won't show the bigger picture. 

Experiment with different numbers of bins. Use `plt.show()` and `plt.clf()` to display and clean up the plot.

### **Customization**  

Creating a plot is one thing, making the correct plot, that makes the message very clear, is the real challenge. There are different types of plot, and for each plot you can do infinite number of customization like color, shape, label, axis and so on. The choice depends on the data and the story you want to tell with this data.

Run again the line plot for population projections till 2100, this time improve it by adding labels to the x and y axes.

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

Add title:

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.show()
```

To put the population growth in perspective, we could have the y-axis started from 0. `plt.yticks()` function does that with the syntax: `plt.yticks([0,1,2], ["one","two","three"])`. The input is a list from 0 to 10 with an interval of 2. The second argument names the ticks (here B stands for billions).

```python
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks([0, 2, 4, 6, 8, 10], ["0", "2B", "4B", "6B", "8B", "10B"])
plt.show()
```
